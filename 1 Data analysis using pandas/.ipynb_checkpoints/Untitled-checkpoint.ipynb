```python
import pandas as pd
```

```python
# list 
data1=[1,2,3,4,5]
pd.DataFrame(data1)
```

```python
data11=pd.DataFrame(data1)
print(data11)
type(data11)
```

```python
data2=[["shan",23],["rihan",43]]
data22=pd.DataFrame(data2)
print(data22)
```

```python
data22=pd.DataFrame(data2,columns=["name","age"])
print(data22)
```

```python
data22=pd.DataFrame(data2,columns=("name","age"),index=[1,2])
print(data22)
```
