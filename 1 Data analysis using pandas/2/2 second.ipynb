```python


# Create Dataset
data = {
    "Name": ["Amit", "Shan", "Ravi", "Amit", "Shan", "Ravi"],
    "Department": ["IT", "HR", "IT", "HR", "IT", "HR"],
    "Salary": [50000, 60000, 55000, 52000, 62000, 58000],
    "Experience": [2, 4, 3, 2, 5, 4]
}

df = pd.DataFrame(data)

print("Original Dataset:\n")
print(df)

# 1. Total Salary by Department
print("\nTotal Salary by Department:")
print(df.groupby("Department")["Salary"].sum())

# 2. Average Salary by Department
print("\nAverage Salary by Department:")
print(df.groupby("Department")["Salary"].mean())

# 3. Multiple Aggregations
print("\nMultiple Aggregations:")
print(df.groupby("Department").agg({
    "Salary": ["sum", "mean", "max"],
    "Experience": "mean"
}))

# 4. Group by Multiple Columns
print("\nSalary by Department and Name:")
print(df.groupby(["Department", "Name"])["Salary"].sum())

# 5. Count Employees in Each Department
print("\nEmployee Count by Department:")
print(df.groupby("Department")["Name"].count())
```

```python
import pandas as pd

# Create Dataset
data = {
    "Department": ["AI", "IT", "AI", "HR", "IT", "AI", "HR"],
    "Salary": [50000, 60000, 55000, 52000, 60000, 50000, 52000],
    "Experience": [2, 4, 3, 2, 5, 2, 2]
}

df = pd.DataFrame(data)

print("Original Data:\n")
print(df)

# 1️⃣ Mean Salary by Department
print("\nMean Salary by Department:")
print(df.groupby("Department").mean())

# 2️⃣ Count of Employees by Department
print("\nCount of Employees by Department:")
print(df.groupby("Department")["Salary"].count())

# 3️⃣ Mode Salary by Department
print("\nMode Salary by Department:")
print(df.groupby("Department")["Salary"].agg(lambda x: x.mode()[0]))
```

# marge  - (column bases ) like sql jion 
inner outer right left


```python
import pandas as pd

students = pd.DataFrame({
    "student_id": [1, 2, 3, 4],
    "name": ["Ali", "Rahul", "Aman", "Sara"]
})

marks = pd.DataFrame({
    "student_id": [1, 2, 4, 5],
    "marks": [85, 90, 78, 88]
})

print(students)
print(marks)
```

```python
pd.merge(students,marks) 
```

```python
pd.merge(students,marks, on="student_id", how="inner") 
```

```python
pd.merge(students,marks, on="student_id", how="outer") 
```

## join - (which is index base)

```python
employees = pd.DataFrame({
    "emp_id": [101, 102, 103, 104],
    "name": ["Amit", "Neha", "Ravi", "Pooja"]
}).set_index("emp_id")

departments = pd.DataFrame({
    "emp_id": [101, 102, 105],
    "department": ["IT", "HR", "Finance"]
}).set_index("emp_id")

print(employees)
print(departments)
```

```python
employees.join(departments)
```

```python

employees.join(departments, how="inner")
# employees.join(departments, how="outer")
```
